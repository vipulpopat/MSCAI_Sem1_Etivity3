```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import math
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('./task4.csv')
X = dataset.values
```

```python
plt.scatter(X[:,0],X[:,1], cmap='seismic')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[:,0],X[:,1],test_size=0.2, random_state=42)
```

```python
plt.scatter(X_train,y_train, cmap='seismic')
plt.show()
plt.scatter(X_test, y_test, cmap='seismic')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
y_train_pred = reg.predict(X_test.reshape(-1, 1))
```

```python
#plotting values 
x_max = np.max(X_test) 
x_min = np.min(X_test) 
y_min = reg.predict(x_min)[0][0]
y_max = reg.predict(x_max)[0][0]

plt.plot([x_min, x_max], [y_min, y_max])
plt.scatter(X_test,y_test, cmap='seismic')
plt.show()
```
